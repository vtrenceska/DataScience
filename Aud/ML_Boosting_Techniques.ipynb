# XGBoost

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/DataScience/Auditoriski/data/bill_authentication.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1:], test_size=0.2)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(max_depth=50, min_child_weight=1, n_estimators=200, learning_rate=0.16)
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import f1_score, classification_report, confusion_matrix
```

```python
print(classification_report(Y_test, y_pred))
```

```python
confusion_matrix(Y_test, y_pred)
```

```python
f1_score(Y_test,y_pred)
```

```python
df = pd.read_csv('/content/drive/MyDrive/DataScience/Auditoriski/data/aud7/train.csv')
```

```python
# df.dropna(axis=1,inplace=True)
```

```python
Y_train
```

```python
df.isnull().sum()
```

```python
df['LotFrontage']
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print(mean_squared_error(Y_test,y_pred))
```

```python
print(r2_score(Y_test,y_pred))
```
