```python
# Initialize an empty list to store all titles
all_titles = []
```

```python
# Extract and append the text of each title to the list
for title in card_titles:
    all_titles.append(title.text)
```

```python
# Print the list of all titles
print(all_titles)
```

```python
all_titles2 = []
for card in card_titles:
  all_titles2.append(card.text)
all_titles2
```

```python
card_times = soup.select('.Card-time')
```

```python
card_times
```

```python
times = []
for card in card_times:
  times.append(card.text)
```

```python
times
```

```python
len(all_titles) == len(times)
```

```python
len(all_titles2) == len(times)
```

```python
from datetime import date
```

```python
times = []
for card in card_times:
  text = card.text
  if 'hour' in text or 'sec' in text or 'min' in text:
    times.append(date.today().strftime("%a, %b %dth %Y"))
  else:
    times.append(card.text)
times
```

```python
card_classes = soup.select('.Card-eyebrow')
card_classes
```

```python
len(all_titles) == len(card_classes)
```

```python
classes = []
for card in card_classes:
  classes.append(card.select_one('div').text)
classes
```

```python
classes = ['Top News', 'Top News', 'Top News', 'Top News', 'Top News'] + classes
```

```python
len(classes)
```

```python
len(all_titles)
```

```python
matrix = []
for i in range(0,len(all_titles)):
  matrix.append([all_titles[i],classes[i],times[i]])
matrix
```

```python
df = pd.DataFrame(np.array(matrix), columns=['title','class','time'])
df
```

```python
matrix = []
for i in range(0,len(all_titles)):
  node={}
  node['title']=all_titles[i]
  node['class']=classes[i]
  node['time']=times[i]
  matrix.append(node)
matrix
```

```python
df = pd.DataFrame(matrix)
df
```
