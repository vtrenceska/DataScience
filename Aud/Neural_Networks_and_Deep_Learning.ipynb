# NN - Keras Model

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Dense
```

```python
# vlezen sloj
model.add(Dense(32,input_dim=13,kernel_initializer='uniform',activation='relu'))
```

```python
# hidden (skrieni) sloevi
model.add(Dense(16,kernel_initializer='uniform',activation='relu'))
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
```

```python
# izlezen sloj
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=100, batch_size=2)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = y_pred.round()
```

```python
from sklearn.metrics import f1_score, confusion_matrix
```

```python
f1_score(Y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(Y_test,y_pred),cmap='Blues',annot=True)
```

# Regression using Neural Networks


```python
df = pd.read_csv('/content/drive/MyDrive/DataScience/Auditoriski/data/Admission_Predict.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df.iloc[:,1:-1],df.iloc[:,-1:],test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential()
```

```python
# vlezen
model.add(Dense(32, input_dim=7, kernel_initializer='normal', activation='relu'))
```

```python
model.add(Dropout(0.2, input_shape=(32,)))
```

```python
model.add(Dense(10, activation='relu'))
```

```python
model.add(Dropout(0.1, input_shape=(10,)))
```

```python
# izlezen
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
history = model.fit(X_train, Y_train, epochs=30, batch_size=32, validation_split=0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train','validation'])
```

```python
y_pred = model.predict(X_test)
```

```python
plt.plot(Y_test, color='blue')
plt.plot(y_pred, color='green')
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,y_pred)
```
